```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("crew_prediction").getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = spark.read.csv("/content/drive/MyDrive/Colab Notebooks/churn_prediction.csv", inferSchema=True, header=True)
```

```python
df.printSchema()
```

```python
df.show()
```

```python
# Print my_spark
print(spark)
# Print the tables in the catalog
print(spark.catalog.listTables())
```

```python
# Creating or replacing a local temporary view with this DataFrame.
df.createOrReplaceTempView("people")
# Define my query
query = "SELECT *, (customer_id*0.3) as bonus_quantity from people"
newdf = spark.sql(query)
#display the content of new dataframe
newdf.show()
```

```python
male = "select * from people where gender = 'Male' "
maledf = spark.sql(male)
maledf.show()
```

```python
Female = "select * from people where gender = 'Female' "
Femaledf = spark.sql(Female)
Femaledf.show()
```
